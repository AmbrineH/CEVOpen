```python
import pandas as pd
from SPARQLWrapper import SPARQLWrapper, JSON
import time
endpoint_url = "https://query.wikidata.org/sparql"
def get_results(endpoint_url, query):
    sparql = SPARQLWrapper(endpoint_url)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()


```

```python
comp = pd.read_table("../oil1000/__tables/compound_multisetCopy.tsv")
```

```python
comp_filt = comp[(comp["WIKIDATA_id"] == "not found")]

comp_filt = comp_filt[comp_filt["PubChem_CID"] != "not found"]
```

Now that the table is filtered, I'm going to fill the original table with the found matches on Wikidata

```python
for index, row in comp_filt.iterrows():
    print("PubChem_CID = " + row["PubChem_CID"])
    pubchem_cid = row["PubChem_CID"]
    sparql_query = '''
    SELECT ?item ?itemLabel 
    WHERE 
    {
      ?item wdt:P662 "''' + pubchem_cid + '''".
      SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
    }
       
    '''
    
    query_failed = False
    
    try:
        results = get_results(endpoint_url, sparql_query)
    except:
        try:
            time.sleep(0.5)
            results = get_results(endpoint_url, sparql_query)
        except:
            try:
                results = get_results(endpoint_url, sparql_query)
            except:
                try:
                    results = get_results(endpoint_url, sparql_query)
                except:
                    query_failed = True
                
            
    if len(results["results"]["bindings"]) > 0 :
    
        wikidata_id = results["results"]["bindings"][0]['item']['value'].split("/")[4]
        wikidata_label = results["results"]["bindings"][0]['itemLabel']['value']
        if wikidata_label == wikidata_id:
            wikidata_label = "No label defined "
        print("Found: " + wikidata_label + "\n QID: "+ wikidata_id)
        
        
        comp.loc[comp.PubChem_CID == pubchem_cid, "WIKIDATA_name"] = wikidata_label
        comp.loc[comp.PubChem_CID == pubchem_cid, "WIKIDATA_id"] = wikidata_id
        
    else:
        print('Not found')

    
    time.sleep(0.5)
    
```
